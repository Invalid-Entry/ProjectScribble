```python
from svglib.svglib import svg2rlg
import svglib
from PIL import Image, ImageDraw
import reportlab
import math
import random
```

```python
def render(coord_paths, size):
    im = Image.new(mode="RGB", size=size)
    d = ImageDraw.Draw(im)
    ImageDraw.floodfill(im,(1,1), (255,255,255))

    for tpath in coord_paths:
        if len(tpath)>0:
            last_point = tpath[0]
            for point in tpath[1:]:
                d.line((last_point,point), fill=(0,0,0))
                last_point = point
    return im
```

```python
# square pos

offset = 250

size = 400

total = size * 4

def coord_calc(point):
    if 0 <= point < size:
        loc = (offset - size//2 + point, offset - size//2)
    elif size <= point < size * 2:
        loc = (offset + size //2 , offset - size//2 + point - size)
    elif size * 2 <= point <  size * 3:
        loc = (offset + (size//2) - (point - 2*size), offset + size//2)
    else:
        loc = (offset - size //2, (offset + size //2) - (point - size * 3) )
        
    return loc
        
paths = []
points = []
p = 1
#points.append(coord_calc(p))
for i in range(100):
    points.append(coord_calc(i))
    tvar = p * 4
    tvar = tvar % (4*size)

    points.append(coord_calc(tvar))
    p += 4
    p = p % (4*size)
        
    point = coord_calc(i)
    #print("%s : %s " % (i, point))
    points.append(point)
    
paths.append(points)
#print(paths)
render(paths, (500,500))
```

```python
paths = []

offset = 250,250
size = 100

steps = 500



angle = 0
step = 0.1

second_angle=0.0
second_step = 0.36
second_size = 50

third_angle=0.0
third_step = 0.12
third_size = 50


path = []

for i in range(steps):
    angle += step
    angle = angle % random.TWOPI
    
    second_angle += second_step
    second_angle = second_angle % random.TWOPI
    
    third_angle += third_step
    third_angle = third_angle % random.TWOPI
    
    x = int(offset[0] + (math.sin(angle)*size) + math.sin(second_angle)*second_size + math.sin(third_angle) * third_size)
    y = int(offset[1] + (math.cos(angle) * size) + math.cos(second_angle)*second_size + math.cos(third_angle) * third_size)
         
    path.append((x,y))

paths.append(path)

render(paths, (500,500))
```
